```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as pPlot
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
removeable_words = ['<br>', 'br', 'Chicago', 'place', 'stay']
```

```python
with open ("review_chicago.txt", 'r') as txt_file:
    data = ""
    for line in txt_file.readlines():
        certain_line = line
        for removeable_word in removeable_words:
            certain_line = certain_line.replace(removeable_word, '')
        data += certain_line

print(data[:100])
```

```python
with open ("review_chicago.txt", 'r') as txt_file:
    print(txt_file.readlines())
```

```python
# 
wordcloud = WordCloud(max_words=100, max_font_size=300 ,background_color="white", width=1600, height=800).generate(data)
plt.figure(figsize=(20,10), facecolor='k')
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
